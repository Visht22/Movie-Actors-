```python
#convert movie release date to datetime data type
df1['movie_release_date'] = pd.to_datetime(df1['movie_release_date'])
```

```python
df1.dtypes
```

```python
#what are the start and end dates of dataset
#start date of movie
df1['movie_release_date'].min()
```

```python
#end date of release year of movie
df1['movie_release_date'].max()
```

```python
#create a new column to store the year of movie release date
df1['movie_release_year']=df1['movie_release_date'].dt.year
```

```python
df1['movie_release_year']
```

```python
df1.columns
```

```python
#creating new columns for unique movie information
df2=df1[['movie_id', 'movie_title', 'movie_popularity', 'movie_release_date',
       'movie_budget', 'movie_revenue','movie_release_year']].drop_duplicates()
```

```python
df2.head()
```

```python
#top ten movies based on movie revenue
df2.nlargest(n=10, columns='movie_revenue')
```

```python
#highest movie revenue movies in each movie_release_year
df2=df2.sort_values(by='movie_revenue')
```

```python
df2.groupby('movie_release_year').last()
```

```python
df2.groupby('movie_release_year').nunique() #no of unique movies for each movie_release_year
```

```python
df2.groupby('movie_release_year')['movie_id'].nunique()
```

```python
df2.groupby('movie_release_year')['movie_revenue'].sum()#total movie revenue for each movie_release_year
```

```python
df2['movie_budget'].hist()
```

```python
sns.jointplot(data=df2, x='movie_budget', y='movie_revenue')
```
